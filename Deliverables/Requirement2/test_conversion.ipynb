```python
import numpy as np
import matplotlib.pyplot as plt

def demand_function(min_price, max_price , curve_steep):

        #curve_steep = np.random.uniform(1, 5)

        conversion_probability = lambda price: np.clip(np.exp(-((np.clip(price, 0, 1) - min_price) / (max_price - min_price)) * curve_steep), 0, 1)

        return conversion_probability

# Example usage (assuming prices are already normalized between 0 and 1)
min_price = 0  # Normalized value
max_price = 1  # Normalized value


prices = np.linspace(0, 1, 100)
epochs = 10
steeps = np.random.uniform(1, 5, epochs)
for epoch in range(epochs):
  demand = demand_function(min_price, max_price, steeps[epoch])
  plt.plot(prices, [demand(price) for price in prices], label=f'Epoch {epoch}')
```
